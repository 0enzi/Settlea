```python
import matplotlib.pyplot as plt
import numpy as np
from collections import namedtuple
```

```python
def plot_hexagon(center, size):
    """Plot a hexagon based on given coord center nd size"""
    hexagon_corners = [get_hex_corners(center, size, i) for i in range(6)]
    hexagon_corners.append(hexagon_corners[0])
    x_coords, y_coords = zip(*hexagon_corners)
    plt.plot(x_coords, y_coords, 'b-')
```

```python
# Parameters
params = {
    'radius': .5,
    'spacing': 1
}

Point = namedtuple('Point', ['x', 'y'])

# Prepare plot
theta = np.linspace(0, 2 * np.pi, 100)
plt.figure(figsize=(10, 10))


coordinates_dict = {}
plot_row(3, params, -0.5, 1, 'row1', coordinates_dict)
plot_row(4, params, -1.5, 0.5, 'row2', coordinates_dict)
plot_row(5, params, -2.5, 0, 'row3', coordinates_dict)
plot_row(4, params, -3.5, 0.5, 'row4', coordinates_dict)
plot_row(3, params, -4.5, 1, 'row5', coordinates_dict)

for row_id, coords in coordinates_dict.items():
    for coord in coords:
        plot_hexagon(coord, .5)

plt.xlim(-0.5, 4.5)
plt.ylim(-5, 0)

plt.grid()
plt.gca().set_aspect('equal', adjustable='box')  # Ensure equal scaling
plt.show()
```

```python
def get_hex_corners(center, size, i):
    angle_deg = 60 * i - 30
    angle_rad = np.pi / 180 * angle_deg
    x = center[0] + size * np.cos(angle_rad)
    y = center[1] + size * np.sin(angle_rad)
    return (x, y)
```

```python
def plot_row(num_items, params, center_y, offset, row_id, coord_dict):
    """Plot a row of circles and store their coordinates."""
    for i in range(num_items):
        center_x = i * params['spacing'] + offset 
#         plt.scatter(center_x, center_y, color='red')  

        # Store the coordinates in the dictionary
        if row_id not in coord_dict:
            coord_dict[row_id] = []
        coord_dict[row_id].append((center_x, center_y))
```
